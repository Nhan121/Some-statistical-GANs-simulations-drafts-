## 2. How to read & display an image?
### 2.1. Using `OpenCV`
#### 2.1.1. Read an image.
$\quad \diamond$ Use the function `cv2.imread()` to read an image; of course, you must import `cv2` before using this function.

$\quad \diamond$ **`Syntax`: `cv2.imread(path, mode_flag)`**, where

$\qquad \qquad \bullet$ The first parameter **`path`** is the `working directory` or a `full path of image` you want to read!

$\qquad \qquad \bullet$ The second parameter **`mode_flag`** is the `flag` which specifies the **`way`** that the image should be read, we have 3 values for this argument:

$\qquad \qquad \quad \circ$ `flag = 1` or `flag = cv2.IMREAD_COLOR`: Loads a `color image`. Any transparency of the image will be neglected! 

$\qquad \qquad \qquad \qquad \qquad \quad \qquad \qquad \qquad \qquad \qquad$ **This is the `default flag`!**

$\qquad \qquad \quad \circ$ `flag = 0` or `flag = cv2.IMREAD_GRAYSCALE`: Loads image in the `gray-scale-mode.`

$\qquad \qquad \quad \circ$ `flag = -1` or `flag = cv2.IMREAD_UNCHANGED`: Loads image as such including `alpha channel.`

**EXAMPLE 1. Load an image with a true-directory (existed file) as a `color image`**

Remember that, when obmiting the second argument (`flag`) then `defalut = 1`

```python
directory = r"C:\Users\Admin\Desktop\Nhan_pro\Data\computer_vision_data\figure_1.jpg"
img_c = cv2.imread(directory)
img_c.shape, type(img_c)
```

- The `type(img_c)` throw `numpy.ndarray` telling that your output is in a form of `array` with 3 dimensions (shape `720x960x3`).
- The `img_c.shape` returns `(581, 774, 3)` meant your image is a color image with: `3 chanels (Red, Green, Blue)`; **`width`** 774 and **`height`** 581! To get more intuitition, look at the function `cv2.imshow()` or `matplotlib.pyplot.imshow()` (in section 2.1.2 this kernel) to understand your output dimension!

**May be you have already know.**
- The values in the `array` of any images fluctuates from `0` to `255`, and they are stored as an `numpy.uint8` or the `8-bit integers` for the pixel value. 
- To get the min_pixel_value, type `image.min()`, also `image.max()` to obtain the max_pixel_value.
- To extract a range from the image, put the range in the `square brackets []` as follow:

                    image[ height_from : height_end, width_from : width_end, : ]

```python
print(img_c.min(), img_c.max())

print(img_c[289:292, 389:392, :])
```

**EXAMPLE 2. Load an image with a true-directory (existed file) as a `gray-scale image`**

```python
img_s = cv2.imread(directory, 0)
img_s.shape, type(img_s)
```

- This time the `img_s.shape` returns `(581, 774)` meant your image is a `2D-array form` and it is not a `color-image`, just a `gray-scale` mode in black and white.

#### 2.1.2. Display an image
**A. Display without using `matplotlib.pyplot`.**

$\qquad \diamond$ **`Syntax`.**

                cv2.imshow(title, img_source)
                cv2.waitKey(0)
                cv2.destroyAllWindows()
                
$\qquad \diamond$ **Code-Explaination.** Reference [here](https://docs.opencv.org/master/d7/dfc/group__highgui.html#ga453d42fe4cb60e5723281a89973ee563)

$\qquad \quad \bullet$ The function `cv2.imshow()` is used to display an image in a window. The window automatically fits to your `image_size`. The first argument `title` is a `string` named the window; while the second is the `image_source`, here is `img_c`.

$\qquad \qquad \qquad$- Of course, you can create as many windows as you want, **but with different `window names`!!**

$\qquad \quad \bullet$ The next function `cv2.waitKey()` is a `keyboard binding function`. Its argument is the time in `milliseconds`. The function waits for specified milliseconds for any keyboard event. If you press any key in that time, the program continues. 

$\qquad \qquad \qquad$- If *`0`* is passed, then it waits indefinitely for a key stroke. 

$\qquad \quad \bullet$ The last function `cv2.destroyAllWindows()` simply destroys all the windows we created. If we want to destroy any specific window; use this function where we pass the exact window_name as the argument.

```python
cv2.imshow('im1_title', img_c)
cv2.waitKey(0)

from IPython.core.display import Image, display
Image(r"C:\Users\Admin\Desktop\fig_1.JPG", width = 530, height = 590)
```

$\qquad$ Above is the **`screenshot`** of the window (with `window_name = "im1_title"`) when using `cv2.imshow()` for the `color image`: `img_c`

$\qquad$ Below, the **`screenshot`** for the `grayscale image`: `img_s` with `window_name = "im0_title"`

```python
cv2.imshow('im1_title', img_s)
cv2.waitKey(0)

Image(r"C:\Users\Admin\Desktop\fig_2.JPG", width = 530, height = 590)
```

**B. Display by using `matplotlib.pyplot`.**
In this case, it required that importing the `matplotlib` library with the `syntax`:

                            import matplotlib.pyplot as plt
Then, to display the image, using

                            plt.figure(figsize = (width, height))
                            plt.imshow(image_source, cmap = 'gray)
                            plt.show()
$\qquad \diamond$ **Code-Explaination.**  

$\qquad \quad \bullet$ The command `plt.figure(figsize = (width, height))` allows you plot or show an image with `(width, height)` in `inches`. If not provided, `defaults` to `rcParams["figure.figsize"] = [6.4, 4.8]`. Read more [here](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.figure.html). 

$\qquad \quad \bullet$ The `plt.imshow(image_source)` is used to display your input-image with `cmap = gray` indicates the black-white scalar.

$\qquad \quad \bullet$ Finally, the `plt.show()` will display all open figures. [Read more](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.show.html)
            
For instance, display the `img_s` by using `matplotlib`

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10, 6))
plt.imshow(img_s, cmap = 'gray')
plt.show()
```

$\diamond$ **Warning:** 

$\qquad \bullet$ The `color image` loaded by `OpenCV` is in BGR (red-green-blue) mode.

$\qquad \bullet$ But in `Matplotlib`, it displays [RGB](https://en.wikipedia.org/wiki/RGB_color_model) mode.

$\qquad \bullet$ Hence, the `color images` will not displayed correctly in `matplotlib` if you used `cv2.imread()`. To load this correctly, we must convert `RGB` to `BGR` and we will discuss this later in the next [track](): [Chapter1_Section1 : changing color_space](https://github.com/Nhan121/Some-statistical-GANs-simulations-drafts-/tree/master/Image_processing%20%26%20computer_vision/Image%20processing/Chapter1).

### 2.2. Using `skimage`
#### 2.2.1. Read an image
**`Syntax`:**

                            skimage.io.imread(path_to_image, as_gray)
where the first argument is the directory to your image, and the second is the option whether the image is `gray` or not (`color`)?

```python
ski_img_c = skimage.io.imread(directory)
ski_img_c.shape, type(ski_img_c)
```

Like `OpenCV`, 
- the `color image` in `skimage` is an `3D-numpy array`.
- the `gray-scale image` is an `2D numpy array`.
- both of these types had the same `size` as in `OpenCV`.

```python
ski_img_s = skimage.io.imread(directory, as_gray = True)
ski_img_s.shape, type(ski_img_s)
```

#### 2.2.2. Display an image.

Unlike in `OpenCV`, the `imshow` function in `skimage.io` didn't need the `name of window`!!

```python
plt.figure(figsize = (20, 8))
plt.subplot(1,2,1), skimage.io.imshow(ski_img_c)
plt.title("color_image with skimage")

plt.subplot(1,2,2), skimage.io.imshow(ski_img_s)
plt.title("gray-scale_image with skimage")
plt.show()
```

$\qquad \diamond$ Likewise, you can also use `matplotlib` to display the image which read from `skimage`

$\qquad \diamond$ Remind that the command `plt.subplot(nrows, ncols, index)` meant the `subplot` will take the `index position` on a `grid` with `nrows` rows and `ncols` columns. 

$\qquad \qquad$ The `index` starts at `1` in the `upper left corner` and `increases` to the `right`. 

$\qquad \qquad$ The `index` can also be a two-tuple specifying the `(first, last)` indices `(1-based, and including last)` of the subplot; e.g., `plt.subplot(3, 1, (1, 2))` makes a `subplot` that spans the `upper 2/3` of the figure.

$\qquad \qquad$ [Read more](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)

```python
plt.figure(figsize = (20, 8))
plt.subplot(1,2,1), plt.imshow(ski_img_c)         ## Display the color_image (ski_img_c) in the left pannel
plt.title("color_image with matplotlib")

plt.subplot(1,2,2), plt.imshow(ski_img_s, cmap = 'gray')
plt.title("gray-scale_image with matplotlib")
plt.show()
```
